ASSIGENMENT#1. Write a program which will find all such numbers which are divisible by 7 but are not amultiple of 5, between 2000 and 3200 (both included). The numbers obtained should be printed in a comma-separated sequence on a single line.

```python
v=[]
for x in range(2000,3200):
    if(x%7==0) and (x%5==0):
        v.append(str(x))
        print (','.join(v))
```
